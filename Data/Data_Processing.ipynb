```python
# Ensure the "tags" column is treated as a string
df["tags"] = df["tags"].astype(str)

# Generate one-hot encoding for tags
tags_dummies = df["tags"].str.get_dummies(sep=", ")

# Select required columns
required_columns = ["imdb_id", "title", "plot_synopsis", "split"]
df_encoded = pd.concat([df[required_columns], tags_dummies], axis=1)
```

```python
# Save the new dataframe to an Excel file
output_file = "encoded_tags.xlsx"
df_encoded.to_excel(output_file, index=False)

print(f"One-hot encoded data saved to {output_file}")
```
